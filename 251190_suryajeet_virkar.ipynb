```python
#Question 1
n = int(input("enter no of elements: "))

a = []
for i in range(n):
  value = int(input("enter elements: "))
  a.append(value)

maximum = a[0]
minimum = a[0]

for num in a:
  if num > maximum:
    maximum = num
  if num < minimum:
    minimum = num
print("maximum value: ",maximum)
print("minimum value: ",minimum)
```

```python
#Question 2
s = input("enter a string: ")
print("you enterd:",s)
left = 0
right = len(s) - 1

is_palindrome = True

while left < right:
  if s[left] != s[right]:
    is_palindrome = False
    break
  left += 1
  right -= 1
if is_palindrome:
  print("s is palindrome")
else:
  print("s is notpalindrome")
```

```python
#Question 3

n = int(input("enter the no of elements: "))
a = []
for i in range(n):
  values = int(input("enter element: " ))
  a.append(values)


x = int(input("enter the value of x: "))
a.sort()
print("sorted array:", a)

i = 0
j = 1
found = False

while j < n and i< n:
  diff = a[j] - a[i]

  if diff == x:
    print("pair found:", a[i],a[j])
    found = True
    break
  elif diff < x:
    j = j  + 1
  else:
    i = i + 1

  if i == j:
    j = j + 1
if not found:
  print("no pair found with difference", x)
```
